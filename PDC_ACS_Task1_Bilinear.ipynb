```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#variable declaration
R=10
C=100e-6
Toe=10e-3 
Ts=0.1e-3
t=[0]
HP=[0]
x=[0]
LP=[1]
I=[0]
h=Ts/Toe
dt=0.1e-3
f=50
w=2*math.pi*f 
u=[0]
v=[1]
```

```python
for n in range(4000):
        t.append(1+t[n])
        u.append(u[n]+w*dt*v[n])
        v.append(v[n]-w*dt*u[n+1])
        x.append(10+u[n+1])
        LP.append(((2-h)/(h+2))*LP[n]+((h/(h+2))*(x[n+1]+x[n])))
        HP.append((2/(h+2))*(x[n+1]-x[n])+((2-h)/(h+2))*HP[n])
        I.append(I[n]+(x[n+1]+x[n])*Ts/2)
```

```python
plt.plot(u, color = 'red') #plotting the values
plt.title('Sine',fontsize=20)
plt.xlabel('Time', fontsize=10)
plt.show()
```

```python
plt.plot(v, color = 'red') #plotting the values
plt.title('Cosine',fontsize=20)
plt.xlabel('Time', fontsize=10)
plt.show()
```

```python
plt.plot(LP, color = 'red') #plotting the values
plt.title('Low Pass Filter',fontsize=20)
plt.xlabel('Time', fontsize=10)
plt.show()
```

```python
plt.plot(HP, color = 'red') #plotting the values
plt.title('High Pass Filter',fontsize=20)
plt.xlabel('Time', fontsize=10)
plt.show()
```

```python
plt.plot(I, color = 'red') #plotting the values
plt.title('Integrator',fontsize=20)
plt.xlabel('Time', fontsize=10)
plt.show()
```
