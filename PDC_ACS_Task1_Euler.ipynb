```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#variable declaration
R=10
C=100e-6
Toe=10e-3 
Ts=0.1e-3
t=[0]
HP=[0]
x=[0]
LP=[0]
I=[0]
h=Ts/Toe
dt=0.1e-3
f=50
w=2*math.pi*f 
u=[0]
v=[1]
```

```python
for n in range(4000):
        t.append(1+t[n])
        u.append(u[n]+w*dt*v[n])
        v.append(v[n]-w*dt*u[n+1])
        x.append(10+u[n+1])
        LP.append(LP[n]+h*(x[n+1]-LP[n])) #Low Pass Filter
        HP.append((1-h)*HP[n]+(x[n+1]-x[n])) #High Pass Filter
        I.append(I[n]+x[n]*Ts)
```

```python
plt.plot(u, color = 'red') #plotting the values
plt.title('Sine',fontsize=20)
plt.xlabel('Time', fontsize=10)
plt.show()
```
